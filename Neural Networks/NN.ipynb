```python
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
df = concrete_data = pd.read_csv('concrete_data.csv')
df.head(10)
# concrete_data.shape
```

```python
labels = df['Strength']
len(labels)
# del df['Strength']
# print(labels)
```

```python
labels.shape
```

```python
concrete_data.head(10)
train=df.sample(frac=0.8,random_state=200)
test=df.drop(train.index)
```

```python
target=labels.sample(frac=0.8,random_state=200)
test_target=df.drop(target.index)
```

```python
test_target
```

```python
# n_cols = df.shape[1]
model.add(Dense(5, activation='relu', input_shape=(df.shape[1],)))
model.add(Dense(5, activation='relu'))
model.add(Dense(1))
```

```python
len(labels)
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(train, target, epochs=200)
```

```python
model.predict(test)
```
